```python
conn = connector.connect(
    host="localhost",
    user="root",
    password="VFFarmy@123",
    database="Student_db"
)       
cursor = conn.cursor()
print("connected:", conn.is_connected())
```

```python
cursor.execute("CREATE TABLE Students (id INT  PRIMARY KEY, name VARCHAR(255), age INT)")
print("Table created successfully")
```

```python
cursor.execute("INSERT INTO Students (id,name, age) VALUES (01,'Bob', 22) ")
print("Record inserted successfully")
```

```python
cursor.execute("select * from Students")
```

```python
sql = "insert into Students (id,name, age) values (%s,%s, %s)"
val = [
    (2,'David', 24),
    (3,'Eva', 25),
    (4,'Frank', 26)
]
cursor.executemany(sql, val)
conn.commit()
print(cursor.rowcount, "records inserted successfully.")
```

```python
rec = cursor.fetchall()
cursor.execute("SELECT * from Students")

for r in rec:
    print(r)
```
